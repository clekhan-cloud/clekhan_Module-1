```python
url="https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/Students_Alcohol_Consumption/student-mat.csv"
df = pd.read_csv(url)
```

```python
df_sliced = df.loc[:, 'school':'guardian']
display(df_sliced.head())
```

```python
capitalize_string = lambda x: str(x).upper()

# Example usage:
text = "hello world"
capitalized_text = capitalize_string(text)
print(f"Original: {text}")
print(f"Capitalized: {capitalized_text}")

text_2 = "python"
print(f"Capitalized: {capitalize_string(text_2)}")
```

```python
df_sliced['Mjob'] = df_sliced['Mjob'].apply(capitalize_string)
df_sliced['Fjob'] = df_sliced['Fjob'].apply(capitalize_string)

display(df_sliced.head())
```

```python
display(df_sliced.tail())
```

```python
df['Mjob'] = df['Mjob'].apply(capitalize_string)
df['Fjob'] = df['Fjob'].apply(capitalize_string)

display(df.head())
```

```python
def majority(age):
    return age > 17

df['legal_drinker'] = df['age'].apply(majority)

display(df.head())
```

```python
df_numeric = df.select_dtypes(include=[np.number]) * 10
df[df_numeric.columns] = df_numeric
display(df.head())
```
